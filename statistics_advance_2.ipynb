Q1.What is hypothesis testing in statistics?

ANS:Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on sample data. It involves testing an assumption (called the **null hypothesis**) against an alternative hypothesis using statistical evidence. The goal is to determine whether there's enough evidence to reject the null hypothesis.


Q2.What is the null hypothesis, and how does it differ from the alternative hypothesis?

ANS:The null hypothesis (H‚ÇÄ) is a statement that there is no effect or no difference, and it represents the status quo or default assumption in a hypothesis test.

The alternative hypothesis (H‚ÇÅ or Ha) is a statement that contradicts the null hypothesis and represents what you aim to prove‚Äîusually that a change, effect, or difference exists.

Q3.> What is the significance level in hypothesis testing, and why is it important8

ANS:The significance level (denoted as Œ±, alpha) in hypothesis testing is the threshold probability used to decide whether to reject the null hypothesis. It represents the maximum acceptable risk of making a Type I error‚Äîthat is, rejecting the null hypothesis when it is actually true.

Q4. What does a P-value represent in hypothesis testing?

ANS:A p-value in hypothesis testing represents the probability of obtaining the observed results (or more extreme) if the null hypothesis is true.

In simple terms:
It measures the evidence against the null hypothesis.

A small p-value (typically ‚â§ 0.05) means the observed data is unlikely under H‚ÇÄ, so we may reject H‚ÇÄ.

A large p-value (> 0.05) means the data is consistent with H‚ÇÄ, so we fail to reject H‚ÇÄ.


Q5. How do you interpret the P-value in hypothesis testing?

ANS:‚úÖ Compare it to the significance level (Œ±):
If p-value ‚â§ Œ±
‚Üí Reject the null hypothesis (H‚ÇÄ)
‚Üí There is statistically significant evidence to support the alternative hypothesis (H‚ÇÅ).

If p-value > Œ±
‚Üí Fail to reject the null hypothesis (H‚ÇÄ)
‚Üí There is not enough evidence to support the alternative hypothesis.

Q6.> What are Type 1 and Type 2 errors in hypothesis testing?

ANS:In hypothesis testing, Type I and Type II errors are the two possible incorrect conclusions:

1. Type I Error (False Positive):
Definition: Rejecting the null hypothesis (H‚ÇÄ) when it is actually true.

Example: Concluding that a new drug works when it actually doesn't.

Probability: Denoted by Œ± (significance level), usually 0.05.

2. Type II Error (False Negative):
Definition: Failing to reject the null hypothesis when the alternative hypothesis (H‚ÇÅ) is actually true.

Example: Concluding that a new drug doesn't work when it actually does.

Probability: Denoted by Œ≤.

Q11. What is the difference between a Z-test and a T-test8?

ANS: A Z-test and a T-test are both statistical methods used to test hypotheses about population means, but they differ based on sample size and the availability of population standard deviation. A Z-test is used when the sample size is large (typically n ‚â• 30) and the population standard deviation (œÉ) is known. It relies on the normal distribution and is ideal for precise and stable estimates. In contrast, a T-test is used when the sample size is small (n < 30) and the population standard deviation is unknown. It uses the t-distribution, which has thicker tails to account for the increased uncertainty in small samples. As the sample size increases, the t-distribution approaches the normal distribution.

Q12.> What is the T-test, and how is it used in hypothesis testing?

ANS:The T-test is a statistical test used to determine whether there is a significant difference between means when the population standard deviation is unknown and/or the sample size is small. It is based on the t-distribution, which adjusts for the extra uncertainty due to estimating the standard deviation from the sample.

In hypothesis testing, the T-test is used to compare:

A sample mean to a known value (one-sample t-test),

The means of two independent groups (independent two-sample t-test), or

The means of paired or related samples (paired t-test).

The test involves calculating a t-score using the sample data, then comparing it to a critical value from the t-distribution table at a chosen significance level (Œ±). If the t-score falls in the critical region, we reject the null hypothesis, suggesting a significant difference exists.

Q20.What is an ANOVA test, and what are its assumptions?

ANS:ANOVA is a statistical method used to compare the means of three or more groups to determine if at least one group mean is significantly different from the others.

It uses the F-distribution to test the null hypothesis that all group means are equal.

Key Assumptions of ANOVA:
Independence:
Observations must be independent within and across groups.

Normality:
Data in each group should be approximately normally distributed.

Homogeneity of Variance (Homoscedasticity):
All groups should have equal variances.

Q21. What are the different types of ANOVA tests?

ANS:There are three main types of ANOVA tests, each suited for different experimental designs and data structures:

1. One-Way ANOVA
Purpose: Compares the means of three or more independent groups based on one factor (independent variable).

Example: Comparing test scores of students from three different teaching methods.

Assumptions: Normality, independence, equal variances.

2. Two-Way ANOVA
Purpose: Compares the means across groups that vary in two factors and also checks for interaction effects between them.

Example: Examining how teaching method and gender affect student scores.

Output: Main effect of factor A, main effect of factor B, and interaction effect (A √ó B).

3. Repeated Measures ANOVA
Purpose: Used when the same subjects are measured multiple times under different conditions or over time.

Q2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range for the discrete uniform distribution
low = 1   # inclusive lower bound
high = 7  # exclusive upper bound (for values 1 to 6, like a die)

# Generate x values (possible outcomes)
x = np.arange(low, high)

# Create a discrete uniform distribution
dist = randint(low, high)

# Get PMF values
pmf_values = dist.pmf(x)

plt.stem(x, pmf_values, basefmt=" ")
plt.xlabel('Value')
plt.ylabel('PMF')
plt.title('PMF of Discrete Uniform Distribution (Die Roll)')
plt.xticks(x)
plt.grid(True)
plt.show()
```

Q3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution>


```python
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): Outcome (0 or 1)
    p (float): Probability of success (0 ‚â§ p ‚â§ 1)

    Returns:
    float: Probability of observing x
    """
    if x not in [0, 1]:
        return 0.0  # Bernoulli outcomes are only 0 or 1
    return p**x * (1 - p)**(1 - x)
```

Q6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range for the discrete uniform distribution (e.g., a fair die)
low = 1
high = 7  # upper bound is exclusive, so values will be 1 through 6

# Create the distribution object
dist = randint(low, high)

# Generate x values (possible outcomes)
x = np.arange(low, high)

# Calculate CDF values
cdf_values = dist.cdf(x)

# Plot the CDF
plt.step(x, cdf_values, where='post', label='CDF')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('CDF of Discrete Uniform Distribution (1 to 6)')
plt.xticks(x)
plt.grid(True)
plt.legend()
plt.show()
```

Q10.> Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Parameters
population_size = 100000
sample_size = 30       # Size of each sample
num_samples = 1000     # Number of samples

# Generate a non-normal population (exponential distribution)
population = np.random.exponential(scale=2.0, size=population_size)

# Collect sample means
sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=False)
    sample_means.append(np.mean(sample))

# Plot the original (non-normal) population distribution
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(population, bins=50, color='orange', edgecolor='black', density=True)
plt.title('Original Non-Normal Population\n(Exponential Distribution)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)

# Plot the sampling distribution of the sample mean
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, color='skyblue', edgecolor='black', density=True)
plt.title(f'Sampling Distribution of Sample Means\n(n = {sample_size}, samples = {num_samples})')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.grid(True)

plt.tight_layout()
plt.show()
```

Q14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    """
    Calculate Z-score for a given data point.
    """
    return (x - mean) / std

# Example data
data_point = 72
population_mean = 70
population_std = 5

# Calculate Z-score
z = calculate_z_score(data_point, population_mean, population_std)
print(f"Z-score of {data_point} is: {z:.2f}")

# Plot standard normal distribution
x_vals = np.linspace(-4, 4, 1000)
y_vals = norm.pdf(x_vals, 0, 1)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label='Standard Normal Distribution', color='blue')
plt.fill_between(x_vals, y_vals, where=(x_vals <= z), color='lightgreen', alpha=0.6, label=f'Z ‚â§ {z:.2f}')
plt.axvline(z, color='red', linestyle='--', label=f'Z-score = {z:.2f}')
plt.title('Z-score Comparison with Standard Normal Distribution')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

Q15. Implement hypothesis testing using Z-statistics for a sample dataset?

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample = np.array([68, 72, 70, 69, 73, 71, 74, 68, 70, 69])
sample_mean = np.mean(sample)
sample_size = len(sample)
sample_std = np.std(sample, ddof=1)  # Sample standard deviation

# Known population mean (H0: mu = 70)
population_mean = 70
alpha = 0.05  # Significance level

# Z-statistic calculation
standard_error = sample_std / np.sqrt(sample_size)
z_score = (sample_mean - population_mean) / standard_error

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")

# Decision
if p_value < alpha:
    print("Reject the null hypothesis (H0)")
else:
    print("Fail to reject the null hypothesis (H0)")
```

Q16. Create a confidence interval for a dataset using Python and interpret the result?

```python
import numpy as np
from scipy.stats import t

# Sample data
data = np.array([52, 55, 48, 50, 53, 49, 54, 51, 56, 47])

# Sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)
confidence = 0.95
alpha = 1 - confidence

# t critical value for two-tailed test
t_critical = t.ppf(1 - alpha/2, df=n-1)

# Margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(n))

# Confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

Q1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results?

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    # Sample statistics
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    # Calculate Z-statistic
    z = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    # Output results
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-statistic: {z:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpret the result
    if p_value < alpha:
        print("Reject the null hypothesis: Sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis: No significant difference from the population mean.")

# Example usage
np.random.seed(42)
sample = np.random.normal(loc=102, scale=15, size=50)  # Simulated sample data

# Perform Z-test
z_test(sample_data=sample, population_mean=100, population_std=15)
```

Q4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test_with_plot(sample_data, population_mean, population_std, alpha=0.05):
    """
    Performs a two-tailed Z-test and visualizes the decision region.

    Parameters:
    sample_data (array-like): The sample data.
    population_mean (float): The hypothesized population mean under the null hypothesis.
    population_std (float): The known population standard deviation.
    alpha (float): The significance level (default is 0.05).
    """
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    # Calculate Z-statistic
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Calculate critical Z-values for a two-tailed test
    z_critical_lower = norm.ppf(alpha / 2)
    z_critical_upper = norm.ppf(1 - alpha / 2)

    # Display results
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Critical Z-values (alpha={alpha}): ({z_critical_lower:.4f}, {z_critical_upper:.4f})")

    # Hypothesis decision
    if abs(z_score) > z_critical_upper: # or p_value < alpha
        print("Reject the null hypothesis: Sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis: No significant difference from the population mean.")

    # --- Visualization ---
    plt.figure(figsize=(10, 6))
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)  # Standard normal distribution

    plt.plot(x, y, label='Standard Normal Distribution', color='blue')

    # Shade the rejection regions
    x_reject_lower = np.linspace(-4, z_critical_lower, 100)
    y_reject_lower = norm.pdf(x_reject_lower, 0, 1)
    plt.fill_between(x_reject_lower, y_reject_lower, color='red', alpha=0.3, label=f'Rejection Region (Œ±/2 = {alpha/2:.2f})')

    x_reject_upper = np.linspace(z_critical_upper, 4, 100)
    y_reject_upper = norm.pdf(x_reject_upper, 0, 1)
    plt.fill_between(x_reject_upper, y_reject_upper, color='red', alpha=0.3)

    # Plot the calculated Z-score
    plt.axvline(z_score, color='green', linestyle='--', label=f'Observed Z-score = {z_score:.2f}')

    # Plot critical values
    plt.axvline(z_critical_lower, color='black', linestyle=':', label=f'Critical Z ({alpha/2:.2f}) = {z_critical_lower:.2f}')
    plt.axvline(z_critical_upper, color='black', linestyle=':', label=f'Critical Z (1-Œ±/2 = {1-alpha/2:.2f}) = {z_critical_upper:.2f}')


    plt.title('Two-Tailed Z-Test Visualization')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example Usage:
np.random.seed(2) # for reproducibility
# Simulate sample data (e.g., testing if the mean differs from 100)
sample_data = np.random.normal(loc=103, scale=15, size=50)

# Perform the two-tailed Z-test and visualize
two_tailed_z_test_with_plot(sample_data, population_mean=100, population_std=15, alpha=0.05)
```

Q5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testingE

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05):
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value for right-tailed test
    z_crit = norm.ppf(1 - alpha)

    # Critical sample mean under H0
    x_crit = mu0 + z_crit * se

    # Type II error: area under H1 to the left of critical value
    beta = norm.cdf((x_crit - mu1) / se)

    # X range for plotting
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # Distributions under H0 and H1
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(x, y0, label='H0: Œº = {:.1f}'.format(mu0), color='blue')
    plt.plot(x, y1, label='H1: Œº = {:.1f}'.format(mu1), color='green')

    # Fill Type I error (alpha)
    plt.fill_between(x, y0, where=(x > x_crit), color='red', alpha=0.4, label='Type I Error (Œ±)')

    # Fill Type II error (beta)
    plt.fill_between(x, y1, where=(x < x_crit), color='orange', alpha=0.4, label='Type II Error (Œ≤)')

    # Annotations
    plt.axvline(x_crit, color='black', linestyle='--', label='Critical Value')

    plt.title("Type I and Type II Error Visualization")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print results
    print(f"Critical Value (xÃÑ): {x_crit:.2f}")
    print(f"Type I Error (Œ±): {alpha}")
    print(f"Type II Error (Œ≤): {beta:.4f}")
    print(f"Power of the test (1 - Œ≤): {1 - beta:.4f}")

# Example usage
visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05)
```

Q9.E Write a Python function to calculate the confidence interval for a sample mean and explain its significance?

```python
import numpy as np
from scipy import stats

def confidence_interval(sample, confidence=0.95):
    """
    Calculate the confidence interval for the sample mean.

    Parameters:
    - sample (array-like): Sample data
    - confidence (float): Confidence level (default 0.95 for 95%)

    Returns:
    - Tuple (lower bound, upper bound)
    """
    n = len(sample)
    mean = np.mean(sample)
    std_err = stats.sem(sample)  # standard error of the mean
    margin = std_err * stats.t.ppf((1 + confidence) / 2, df=n-1)

    return mean - margin, mean + margin

# Example usage
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=30)
ci_low, ci_high = confidence_interval(data, confidence=0.95)

print(f"Sample Mean: {np.mean(data):.2f}")
print(f"95% Confidence Interval: ({ci_low:.2f}, {ci_high:.2f})")
```

Q10. Write a Python program to calculate the margin of error for a given confidence level using sample data?

```python
import numpy as np
from scipy import stats

def margin_of_error(sample, confidence=0.95):
    """
    Calculate the margin of error for a sample mean.

    Parameters:
    - sample (array-like): The sample data
    - confidence (float): The confidence level (default = 0.95)

    Returns:
    - margin of error (float)
    """
    n = len(sample)
    std_err = stats.sem(sample)  # standard error of the mean
    t_score = stats.t.ppf((1 + confidence) / 2, df=n-1)  # critical t-score
    return t_score * std_err

# Example usage
np.random.seed(0)
sample_data = np.random.normal(loc=70, scale=12, size=50)

moe = margin_of_error(sample_data, confidence=0.95)
mean = np.mean(sample_data)

print(f"Sample Mean: {mean:.2f}")
print(f"Margin of Error (95% CI): ¬±{moe:.2f}")
print(f"Confidence Interval: ({mean - moe:.2f}, {mean + moe:.2f})")
```

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Create a contingency table
data = np.array([[30, 10],   # Male
                 [20, 20]])  # Female

# Step 2: Perform Chi-Square Test
chi2, p, dof, expected = chi2_contingency(data)

# Step 3: Print results
print("Chi-square Statistic:", round(chi2, 4))
print("Degrees of Freedom:", dof)
print("P-value:", round(p, 4))
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, columns=["Likes", "Dislikes"], index=["Male", "Female"]))

# Step 4: Interpretation
alpha = 0.05
if p < alpha:
    print("\nüîç Result: Reject the null hypothesis (variables are dependent)")
else:
    print("\n‚úÖ Result: Fail to reject the null hypothesis (variables are independent)")
```

Q14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.


```python
from scipy.stats import chisquare

# Step 1: Define observed and expected frequencies
observed = [8, 9, 10, 12, 11, 10]
expected = [10, 10, 10, 10, 10, 10]

# Step 2: Perform the Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Step 3: Output results
print("Chi-square Statistic:", round(chi2_stat, 4))
print("P-value:", round(p_value, 4))

# Step 4: Interpretation
alpha = 0.05
if p_value < alpha:
    print("‚ùå Reject the null hypothesis ‚Äî observed distribution differs from expected.")
else:
    print("‚úÖ Fail to reject the null hypothesis ‚Äî observed distribution fits expected.")
```
